```python
from pandas import read_csv
from pathlib import Path

data = read_csv(
    Path("../data/eistage.csv"),
    nrows=70,
    skiprows=2,
    names=["year", "heat", "ice"],
    usecols=[3, 5, 6],
    sep=";",
)

data.dropna(inplace=True)

data["extreme_days"] = data["heat"] + data["ice"]

print(data.head())
```
